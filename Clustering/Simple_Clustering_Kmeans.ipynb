```python
# Langkah 1: Import library
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

image.png

```python
# Langkah 2: Siapkan data
data = {
    'Sample Data': [1, 2, 3, 4, 5, 6],
    'X': [100, 40, 30, 90, 65, 25],
    'Y': [50, 60, 70, 10, 40, 35]
}
df = pd.DataFrame(data)

plt.show(data)
```

```python
# Langkah 3: Pilih kolom yang ingin di-cluster
X = df[['X', 'Y']]

```

```python
# Langkah 4: Jalankan K-Means Clustering
kmeans = KMeans(n_clusters=2, random_state=42)  # Menggunakan 2 cluster sebagai contoh
df['Cluster'] = kmeans.fit_predict(X)


```

```python
# Langkah 5: Tampilkan hasil clustering
print(df)

# Visualisasi hasil clustering (opsional)
plt.scatter(df['X'], df['Y'], c=df['Cluster'], cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X')  # Centroid
plt.title('K-Means Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
